```python
# Importing necessary libraries
import warnings
warnings.simplefilter('ignore', FutureWarning)
import pandas as pd
import numpy as np
```

```python
# Reading data from an Excel file into a Pandas DataFrame
data1 = pd.read_excel('Murder Cases (State_UT-wise) - 2019-2021.xlsx', skiprows = 2)
data1
```

```python
# Dropping specific rows from the DataFrame
rows = [0, 1, 30, 31, 40, 41, 42, 43, 44, 45, 46, 47 ]
data1.drop(index=rows, inplace = True)
data1
```

```python
data1.columns
```

```python
# Selecting specific columns in the DataFrame
Columns = ['State/UT', 'Mid-Year Projected Population (in Lakhs) (2021)', 'Rate of Murder (2021)', 'Chargesheeting Rate (2021)' ]
data1 = data1[Columns]
data1
```

```python
# Setting 'State/UT' column as the index of the DataFrame
data1.set_index('State/UT', inplace=True)

# Renaming specific index values in the DataFrame
data1.rename(index={'D&N Haveli and Daman & Diu+': 'D&N Haveli and Daman & Diu', 
                    'Jammu & Kashmir*': 'Jammu & Kashmir', 
                    'Ladakh @': 'Ladakh'}, inplace=True)
```

```python
data1
```

```python
data1.columns
```

```python
# Resetting the index of the DataFrame
data1.reset_index(inplace=True)

# Updating the index starting from 1
data1.index = range(1, len(data1) + 1)
data1
```

### Reading Second Dataset

```python
# Reading data from an Excel file into a Pandas DataFrame

data2 = pd.read_excel('Victims of Murder (Gender & Age Group-wise) - 2021.xlsx', skiprows =6)
data2
```

```python
# Dropping specific rows from the DataFrame
rows2 = [28,29,38,39,40,41,42,43,44,45]
data2.drop(index = rows2, inplace = True)
data2
```

```python
data2.columns
```

```python
# Removing specific columns from the DataFrame
Columnstoremove = ['STATES:.1','Unnamed: 15','STATES:.2', 'Unnamed: 29','STATES:.3', 'Unnamed: 43']
data2.drop(columns = Columnstoremove, inplace = True)
data2
```

```python
data2.shape
```

```python
# Renaming a specific column in the DataFrame
data2.rename(columns={'Unnamed: 1' : 'State/UT'}, inplace = True)
data2
```

```python
# Specifying column names for the DataFrame

Columnsname = ['SL','State/UT',
'Child (Below 6 Years) M','Child (Below 6 Years) F','Child (Below 6 Years) Trans','Child (Below 6 Years) Total',
'Child (6 to 12 Years) M','Child (6 to 12 Years) F','Child (6 to 12 Years) Trans','Child (6 to 12 Years) Total',
'Child (12 to 16 Years) M','Child (12 to 16 Years) F','Child (12 to 16 Years) Trans','Child (12 to 16 Years) Total',
'Child (16 to 18 Years) M','Child (16 to 18 Years) F','Child (16 to 18 Years) Trans','Child (16 to 18 Years) Total',
'Total Child Victims M','Total Child Victims F','Total Child Victims Trans','Total Child Victims',
'Adult (18 to 30 Years) M','Adult (18 to 30 Years) F','Adult (18 to 30 Years) Trans','Adult (18 to 30 Years) Total',
'Adult (30 to 45 Years) M','Adult (30 to 45 Years) F','Adult (30 to 45 Years) Trans','Adult (30 to 45 Years) Total',
'Adult (45 to 60 Years) M','Adult (45 to 60 Years) F','Adult (45 to 60 Years) Trans','Adult (45 to 60 Years) Total',
'Adult (Above 60 Years) M','Adult (Above 60 Years) F','Adult (Above 60 Years) Trans','Adult (Above 60 Years) Total',
'Total Adult Victims M','Total Adult Victims F','Total Adult Victims Trans','Total Adult Victims',
'Total Victims M','Total Victims F','Total Victims Trans','Total Victims']
```

```python
# Assigning specified column names to the DataFrame
data2.columns = Columnsname
data2
```

```python
# Removing specific columns from the DataFrame
data2.drop(columns='SL', inplace=True)
data2
```

```python
# Merging two DataFrames based on 'State/UT' column using outer join

data2021 = pd.merge(data1, data2, on = 'State/UT', how = 'outer')
data2021
```

```python
data2021.columns
```

```python
# Saving the merged DataFrame to a CSV file

data2021.to_csv('Murder_Case(State Wise)_2021', index = False)
```
