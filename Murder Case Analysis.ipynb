### Transforming The Data

#### The Rate of Murder from 2020 - 2022 is incorrect. To make the data easier and correct with some functions, using Total Number Of Murders and Total Population.

#### Murder Rate = ( Total Number Of Murders / Total Population In Lakhs ) × 100,000

```python
#Define the function to calculate the murder rate.

def calculate_murder_rate(population_lakhs, total_murders):
    population = population_lakhs * 100000  # Convert lakhs to individuals
    if population == 0:  # To avoid division by zero
        return 0
    murder_rate = (total_murders / population) * 100000
    return round(murder_rate, 2)  # Round to 2 decimal places
```

```python
#Apply the function to update the dataframe.
#Lambda Function with 'apply'.

df["Rate of Murder (2022)"] = df.apply(lambda row: calculate_murder_rate(row["Mid-Year Projected Population (in Lakhs) (2022)"], row["Total Number Of Murders(2022)"]), axis=1)

df.head(10)
```

```python
#Apply the function to update the dataframe.
#Lambda Function with 'apply'.
df2['Rate of Murder (2021)'] = df2.apply(lambda row: calculate_murder_rate(row["Mid-Year Projected Population (in Lakhs) (2021)"], row["Total Number Of Murders(2021)"]), axis = 1)

df2.head(10)
```

```python
#Apply the function to update the dataframe.
#Lambda Function with 'apply'.

df3['Rate of Murder (2020)'] = df3.apply(lambda row: calculate_murder_rate(row["Mid-Year Projected Population (in Lakhs) (2020)"], row["Total Number Of Murders(2020)"]), axis = 1)

df3.head(10)
```

## Distribution of Highest Murders in Adult Category by State/UT (2020-2022)

### Description:

This separate bar plots for each year (2020, 2021, 2022), illustrating the distribution of the highest murders in various Adult category subgroups across different states or union territories (State/UT). Each plot highlights the category within the Adult group that recorded the maximum murders for that particular year. The x-axis displays the State/UT names, while the y-axis represents the total number of murders. The visualization aims to provide insights into the geographical distribution of severe adult-related crimes over the specified years.

```python
years = ['2020', '2021', '2022']
dfs = [df3, df2, df]  # Replace with your actual DataFrames

# Initialize variables to store max values and corresponding categories
max_values = []
max_categories = []

# Find max values and corresponding categories for each year
for year, data in zip(years, dfs):
    max_category = data[columns_of_adults].sum().idxmax()  # Find category with max sum of murders
    max_value = data[max_category].max()
    max_categories.append((max_category, max_value))
    max_values.append(max_value)
```

```python
# Plotting max category against State/UT for each year separately
for i, (year, data) in enumerate(zip(years, dfs), start=1):
    plt.figure(figsize=(12, 6))
    
    # Plot bar chart for the current year
    bars = plt.bar(data['State/UT'], data[max_categories[i-1][0]], color='tan', alpha=0.8)
    
    # Highlight state with maximum murders
    max_state_index = data[max_categories[i-1][0]].idxmax()
    bars[max_state_index].set_color('orangered')  # Highlight max state in orangered
    
    plt.title(f'Maximum Murders Category ({max_categories[i-1][0]}) in {year}')
    plt.xlabel('State/UT')
    plt.ylabel('Total Murders')
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.grid(True)
    
   #show Plot
    plt.show()
```
